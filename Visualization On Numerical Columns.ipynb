# Visualization

## Line PLot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(50,20))
plt.plot(df['Quantity'], label='Quantity')
plt.title('Line plot of Quantity')
plt.xlabel('Index')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Sales'],df['Discount'],c='green',alpha=0.5)
plt.title('Sactter Plot of Sales vs Discount ')
plt.xlabel('Sales')
plt.ylabel('Discount')
plt.show()
```

## Bar Plot

```python
plt.figure(figsize=(10,5))
plt.bar(df.index[:50],df['Sales'][:50],color='blue')
plt.title('Bar Plot of Sales (First 50 rows)')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.show()
```

```python
df.columns
```

## Histogram

```python
plt.figure(figsize=(10,5))
plt.hist(df['Sales'],bins=20,color='Green',alpha=0.7)
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('index')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Quantity'], bins=10,color='Pink')
plt.title('Histrogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Index')
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10,5))
plt.boxplot([df['Sales'],df['Profit'],df['Discount'], df['Shipping Cost'],df['Quantity']], labels=['Sales','Profit','Discount','Shipping Cost','Quantity'])
plt.title('Box plot of sales,profit,discount,shipping cost and quantity')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['Sales'], labels=['Sales'])
plt.title('Box plot of sales')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['Profit'], labels=['Profit'])
plt.title('Box plot of profit')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['Discount'], labels=['Discount'])
plt.title('Box plot of sales,profit,discount,shipping cost and quantity')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot( df['Shipping Cost'], labels=['Shipping Cost'])
plt.title('Box plot of sales,profit,discount,shipping cost and quantity')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['Quantity'], labels=['Quantity'])
plt.title('quantity')
plt.ylabel('Value')
plt.show()
```

## Pie Chart

```python
market_profit = df.groupby('Market')['Profit'].sum()

# Pie chart for Profit by Market
plt.figure(figsize=(6, 6))
plt.pie(market_profit, labels=market_profit.index, autopct='%1.1f%%')
plt.title('Pie Chart of Profit by Market')
plt.show()
```

```python
df.columns
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

# Pie chart for Sales by Category
plt.figure(figsize=(8, 8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Sales Across Categories')
```
